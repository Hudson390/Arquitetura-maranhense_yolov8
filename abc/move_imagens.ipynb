```python
import os
import shutil
import random

# Define folder paths
source_folder = r'D:\Documentos\GitHub\Arquitetura-maranhense_yolov8\abc'  # Replace with your folder path
validation_folder = r'D:\Documentos\GitHub\Arquitetura-maranhense_yolov8\abc\val'  # Replace with your validation folder path

# Create validation folder if it doesn't exist
if not os.path.exists(validation_folder):
    os.makedirs(validation_folder)

# List all images in the source folder (assuming images have common extensions like .jpg, .png)
images = [f for f in os.listdir(source_folder) if f.endswith(('.jpg', '.png', '.jpeg'))]

# Shuffle and select 30 random images
random.shuffle(images)
selected_images = images[:30]

# Move selected images and their corresponding .txt files to the validation folder
for image in selected_images:
    source_path_image = os.path.join(source_folder, image)
    destination_path_image = os.path.join(validation_folder, image)
    shutil.move(source_path_image, destination_path_image)
    
    # Move corresponding .txt file if exists
    txt_file = image.replace(os.path.splitext(image)[1], '.txt')
    source_path_txt = os.path.join(source_folder, txt_file)
    if os.path.exists(source_path_txt):
        destination_path_txt = os.path.join(validation_folder, txt_file)
        shutil.move(source_path_txt, destination_path_txt)

print(f"Moved {len(selected_images)} images and their corresponding .txt files to {validation_folder}")
```
